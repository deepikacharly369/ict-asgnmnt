```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

#1.Reading the given dataset into python environment using pandas

data=pd.read_excel('iris.xls')
print(data.to_string())
```

```python
#2. Displaying the columns in the dataset

data.columns
```

```python
data.shape
```

```python
#3.Calculating the mean of each column of the dataset.

avg=data.mean(numeric_only=True)
print(avg)
```

```python
#4. Looking for missing values

data.isnull()
```

```python
#4.Count of missing values 

missing_values=data.isna().sum()
print(missing_values)
```

```python
#5.meaningful visualizations using the dataset.

sns.countplot(x=data.Classification) 
plt.show()
```

```python
#There are 3 classifications each with a count of 50
```

```python
#scatter plots relating SL&SW and PL&PW respectively.

sns.scatterplot(x=data["SL"],y=data["SW"], hue=data["Classification"])
plt.show()
```

```python
sns.scatterplot(x=data["PL"],y=data["PW"], hue=data["Classification"])
plt.show()
```

```python
#from the above 2 plots it is noticeable that using petal length and petal width rather than sepal length and sepal width is more helpful while classification( in the SL&SW scatter plot versicolar and virginica classes overlaps and it is hard to distinquish between the two eventhough setosa remains distinquishable)
```

```python
#Box plot

sns.boxplot(x = 'Classification', y='PL', data=data)
plt.show()
```
